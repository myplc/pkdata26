```python
# ì˜ˆì œ4 ìº¡ìŠí™”
class Bank:
    __interest_rate = 0.02 # í´ë˜ìŠ¤ ë³€ìˆ˜ p210
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance or 0
    def bprint(self): # ì”ì•¡ì¡°íšŒ
        return f"{self.owner }ë‹˜ì˜ ì”ê³ ëŠ” {self.__balance}ì› ì…ë‹ˆë‹¤."
    def deposit(self, amount): # ì…ê¸ˆ
        self.__balance += amount
        return f"{self.owner }ë‹˜ì˜ ê³„ì¢Œì— {amount}ì›ì´ ì…ê¸ˆ ë˜ì—ˆìŠµë‹ˆë‹¤. í˜„ì¬ì”ì•¡: {self.__balance}ì›"
    def withdraw(self, amount): # ì¶œê¸ˆ
        if self.__balance >= amount:
            self.__balance -= amount
            return f"{self.owner }ë‹˜ì˜ ê³„ì¢Œì—ì„œ {amount}ì›ì´ ì¶œê¸ˆ ë˜ì—ˆìŠµë‹ˆë‹¤. í˜„ì¬ì”ì•¡: {self.__balance}ì›"
        return "ì”ì•¡ì´ ë¶€ì¡±í•©ë‹ˆë‹¤."
    def interest(self):
        self.__balance = self.__balance * (1 + self.__interest_rate)
        return f"{self.__interest_rate}% ì˜ ì´ìê°€ ë°˜ì˜ë˜ì–´ í˜„ì¬ì”ì•¡: {self.__balance:.2f}ì›"
```

```python
# ì˜ˆì œ5 ì‘ìš©
import random
class Character:
    def __init__(self, name, hp, attack_p):
        """ì¼€ë¦­í„° ì´ˆê¸°í™”"""
        self.name = name
        self.hp = hp
        self.attack_p = attack_p
    def attack(self, opponent):
        """ìƒëŒ€ë°©ì„ ê³µê²©"""
        damage = random.randint(2, self.attack_p)
        opponent.hp -= damage
        print(f"{self.name}ê°€ {opponent.name}ì„ ê³µê²©í•˜ì—¬ {damage}ë¥¼ ì…í˜”ìŠµë‹ˆë‹¤.")
    def now_status(self):
        """í˜„ì¬ ì²´ë ¥ ìƒíƒœ í™•ì¸"""
        return f"ì´ë¦„:{self.name} ì²´ë ¥:{self.hp}"
    def is_alive(self):
        """ì¼€ë¦­í„° ìƒì¡´ì—¬ë¶€ í™•ì¸"""
        return self.hp > 0   
```
